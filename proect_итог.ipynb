```python
def categorize_purpose(purpose):
    if 'автомобиль' in purpose or 'автомобиля' in purpose or 'автомобили' in purpose or 'автомобилем' in purpose:
        return 'операции с автомобилем'
    elif 'недвижимость' in purpose or 'жилье' in purpose or 'жилья' in purpose or 'жилью' in purpose or 'жильем' in purpose or 'недвижимости' in purpose or 'недвижимостью' in purpose:
        return 'операции с недвижимостью'
    elif 'свадьба' in purpose or 'свадьбу' in purpose or 'свадьбы' in purpose:
        return 'проведение свадьбы'
    elif 'образование' in purpose or 'образованием' in purpose or 'образования' in purpose:
        return 'получение образования' 
```

```python
df['purpose_category'] = df['purpose'].apply(categorize_purpose)
```

```python
dolg_ot_det = pd.pivot_table(df, index='children', values='debt', aggfunc=['count', 'sum'])
dolg_ot_det.columns = ['Всего кредитополучателей', 'Всего должников']
dolg_ot_det['Доля должников'] = (dolg_ot_det['Всего должников'] / dolg_ot_det['Всего кредитополучателей']) * 100
dolg_ot_det
```

```python
dolg_ot_det['Доля должников'].plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title('Процент должников по количеству детей')
plt.xlabel('Количество детей')
plt.ylabel('Процент должников')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
